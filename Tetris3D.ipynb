```python
# 1. Instalação das bibliotecas
!pip install -q -U google-generativeai

import google.generativeai as genai

# 2. Configuração da API (Substitua pela sua Key)
GOOGLE_API_KEY = "AIzaSyCK8PA16x9KfzFPD-T1FADEiKHovwvjF1E"
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-2.5-flash')

# 3. Motor de Decisão (Sistema Especialista)
def motor_decisao(distancia, obstaculo, alinhado):
    if obstaculo:
        return "PARADA_EMERGENCIA", "Obstáculo detectado na rota de colisão."
    if not alinhado:
        return "CORRECAO_TRAJETORIA", "AGV desalinhado em relação ao eixo do caminhão."
    if distancia > 0.5:
        return "AVANCAR", "Caminho livre. Aproximando-se da zona de carga."
    else:
        return "CARREGAR", "Posição ideal alcançada. Iniciando operação."

# 4. Simulação de Dados (Inputs do LiDAR/Sensores)
dist_lidar = 0.3  # metros
sensor_obstaculo = False
esta_alinhado = True

decisao_tecnica, motivo = motor_decisao(dist_lidar, sensor_obstaculo, esta_alinhado)

# 5. Integração com Gemini (Análise Interpretativa)
prompt = f"""
O sistema especialista de um AGV de carregamento de caminhões tomou a seguinte decisão: {decisao_tecnica}.
Motivo técnico: {motivo}.
Contexto: Distância de {dist_lidar}m, Obstáculo: {sensor_obstaculo}, Alinhado: {esta_alinhado}.

Explique para um operador humano o porquê desta decisão e dê uma recomendação de segurança.
"""

response = model.generate_content(prompt)

print(f"--- OUTPUT DO SISTEMA ---")
print(f"Decisão: {decisao_tecnica}")
print(f"\n--- ANÁLISE DO GEMINI ---")
print(response.text)
```
